```python
#Crea columnes noves (velocitat mitjana del vol)
Airlines_Delay[['DepTime','ArrTime','ArrDelay','AirTime','ActualElapsedTime', 'CRSElapsedTime','TaxiIn','TaxiOut']]
```

```python
#velocitat mitjana del vol='AvFlightSpeed'
Airlines_Delay['AvFlightSpeed'] = Airlines_Delay['Distance'] / Airlines_Delay['AirTime'] * 60
Airlines_Delay[['Distance','AirTime','AvFlightSpeed']]
```

```python
#Crea columnes noves (si ha arribat tard o no)
def Flight_status_function(x):
    if x > 0:
        return "Late"
    else:
        return "On time"
Airlines_Delay["Flight_status"] = Airlines_Delay["ArrDelay"].apply(Flight_status_function)
Airlines_Delay[['DepTime','ArrDelay','Flight_status']]
```

```python
#Taula de les aerolínies amb més endarreriments acumulats
```

```python
#Seleccioanr only 'Late' flight
Airlines_Delay_Late = Airlines_Delay[Airlines_Delay['Flight_status']=='Late']
print(Airlines_Delay_Late)
```

```python
#Seleccionar columnas para tabla reducida
Worst_Airlines_Delay = Airlines_Delay_Late[['UniqueCarrier','ArrDelay','Flight_status']]
print(Worst_Airlines_Delay)
```

```python
#Count endarreriments acumulats per aerolínies
Result=Worst_Airlines_Delay.set_index(['UniqueCarrier','Flight_status']).count(level='UniqueCarrier')
print(Result)
```

```python
#Sprint3_Tasca6
#visualització per: Una variable categòrica (UniqueCarrier)
Result.plot.bar(title = "Number of Flight delays by Carrier", color = "red")
```
